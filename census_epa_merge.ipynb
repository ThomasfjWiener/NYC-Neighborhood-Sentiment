```python
import pandas as pd
```

```python
census_df = pd.read_csv('github_data/census_for_merge.csv')
EPA_df = pd.read_csv('github_data/EPA_df_for_analysis.csv')
```

```python
# need to update unique identifier for EPA_df
EPA_df['STATE'] = '36'
EPA_df['STATE'] = EPA_df['STATE'].astype(str)
EPA_df['TRACTCE'] = EPA_df['TRACTCE'].astype(str).str.zfill(6)
EPA_df['COUNTYFP'] = EPA_df['COUNTYFP'].astype(str).str.zfill(3)
EPA_df['BLKGRPCE'] = EPA_df['BLKGRPCE'].astype(str)

# Create a unique identifier by concatenating the formatted strings
EPA_df['CBG_ID'] = EPA_df['STATE'] + EPA_df['COUNTYFP'] + EPA_df['TRACTCE'] + EPA_df['BLKGRPCE']
```

```python
print(census_df[['CBG_ID']].dtypes)
print(EPA_df[['CBG_ID']].dtypes)
```

```python
census_df['CBG_ID'] = census_df['CBG_ID'].astype(str)
EPA_df['CBG_ID'] = EPA_df['CBG_ID'].astype(str)
```

```python
EPA_df[['CBG_ID','STATE','TRACTCE','COUNTYFP','BLKGRPCE']].head()
```

Merge time

```python
merged_df = pd.merge(census_df, EPA_df, on='CBG_ID', how='inner')
```

```python
print(len(merged_df))
print(merged_df.dtypes)
```

```python
merged_df['TRACT_ID'] = merged_df['CBG_ID'].str[:11]

merged_df[['CBG_ID','TRACT_ID']].head()
```

```python
merged_df.head()
```

```python
merged_df.drop(columns=['state', 'county', 'Total Population', 'tract', 'block group', 'STATE'], inplace=True)
print(merged_df.dtypes)
```

```python
merged_df.to_csv('github_data/merged_data_for_analysis.csv', index=False)
```
